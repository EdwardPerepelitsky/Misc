```python
import numpy as np
from scipy import stats as stats
```

```python
num_trials = [0,2,10,20,50,500]
```

```python
data = stats.bernoulli.rvs(0.5,size = num_trials[-1])
```

```python
xvals = np.linspace(0,1,100)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,7))
for i, N in enumerate(num_trials):
    num_heads = np.sum(data[:N])
    density = stats.beta.pdf(xvals,num_heads+1,N-num_heads+1)
    plt.subplot(2,3,i+1)
    plt.xlabel("P(H)")
    plt.ylabel("P")
    plt.plot(xvals,density)
plt.show()
    
```
